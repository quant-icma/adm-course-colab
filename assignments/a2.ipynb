 $$N_2(u,v,\rho) = \frac{1}{2 \pi \sqrt{1-\rho^2}}\int_{-\infty}^u \int_{-\infty}^v \exp\left( - \frac{1}{2} \left(\frac{x^2 - 2 \rho x y + y^2}{1 - \rho^2}\right) \right) dx dy,$$
