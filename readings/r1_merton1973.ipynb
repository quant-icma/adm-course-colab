## Notation

The paper uses mostly arbitrage reasoning to deduce constraints that option pricing
formulae should satisfy in an arbitrage-free market. 

Using the notation

$c$: Eureopean call option price

$C$: American call option price

$p$: European put option price

$P$: European put option price

$S_0$: Current stock price

$B_0$: Current price of zero-coupon bond of maturity $T$

$S_T$: Stock price at time $T$

$K$: Option strike price

$T$: Option time-to-expiry

$\sigma$: Stock volatility

$r$: Continuously compounding risk-free rate

## Positive payoffs imply positive prices

At expiry, the holder of an option would only exercise if the payoff is positive. 
When exercising a call, the holder pays $K$ and receives the underlying worth $S_T$.
When exercising a put, the holder receives $K$ in exchange of the underlying worth $S_T$.

Thus, at expiry the payoff functions for the call and put are
$$\max\left[S_T - K, 0\right]$$
and
$$\max\left[K - S_T, 0\right]$$
respectively.

At expiry, since the option can be exercised into the payoff, then the value should match the payoff, and so we have
$$C = c = \max\left[S_T - K, 0\right]$$
and 
$$P = p = \max\left[K - S_T, 0\right]$$

Since the option payoffs are always positive or zero, their prices should be too:
$$C \geq 0$$
$$c \geq 0$$
$$P \geq 0$$
$$p \geq 0$$

## American options are worth more than their intrinsic value

Because an American option can be exercised at any time until expiration, then at any time before
expiry they should be worth at least as much as the intrinsic value from immediate exercise
$$C \geq \max\left[S - K, 0\right]$$
and 
$$P \geq \max\left[K - S, 0\right]$$
Otherwise, there would be an arbitrage opportunity by buying the option and exercising immediately
realizing a profit. This is not necessarily the case with European options since they can only be exercised
at expiry.

## Optionality is valuable

Given two identical American options except for one having a longer time-to-expiration, then the longer
maturity option is worth at least as much as the shorter maturity option, since whenever the shorter
maturity option can be exercised, so can the longer maturity option. Hence, as an American option price is an
increasing function of time-to-expiry. This is not necessarily the case with European options since the option
may be so much in-the-money that it can only loose value while waiting for expiry to be able to exercise.

## American options are more valuable than Europeans

By the same reasoning, an American option is worth at least as much as the equivalent European option.
$$C \geq c$$
$$P \geq p$$

## Calls are decreasing in strike, Puts are increasing in strike

Because the holder of a call option has to pay the strike at exercise, all else being equal a call option price
is a decreasing function of the strike. Inversely, the holder of a put option receives the strike at exercise and hence a put option price is an increasing function of the strike. This holds for American and European options.

## European call-put parity

If we hold a European call option with strike $K$ and sell a European put option with strike $K$, then our payoff will
be 
$$\max(S-K,0)-\max(K-S,0)=S-K$$
This is the same payoff as a forward contract with forward price $K$, so the portfolio of long call and short put 
should be worth the same as the forward contract
$$c-p = S-K B_t$$

## Calls are bounded above by the stock, Puts are bounded above by the strike

Given the payoff of a call option is lower than the value of the underlying stock
$$S_T \geq \max\left[S_T - K,0\right]$$
it has to be that the value of the call option is dominated by the value of the stock at any time
$$S_t \geq c_t$$
$$S_t \geq C_t$$

Similarly, the payoff of a put is lower than the strike
$$K \geq \max\left[K - S_T,0\right]$$
thus the put is dominated by the strike at any time
$$K \geq p_t$$
$$K \geq P_t$$

Given the call-put parity relationship $c-p = S-K B_t$, we have
$$c-S = p-K B_t$$
and since $c-S\leq 0$ then we have the tighter bound for European puts
$$KB_t \geq p_t$$

## Calls are bounded below by the intrinsic value

So far we have shown that for American calls we have bounded the option price from both sides
$$ \max\left[S - K, 0\right] \leq C \leq S$$
The upper bound also holds for European calls.

Now, we show a lower bound for a European call on a stock that pays no dividends by considering a portfolio of one option and $K$ zero-coupon bonds with maturity $T$. At the option expiry, the zero-coupon bonds pay $K$ and the option pays $\max(S-K,0)$ 
and thus the payoff of the portfolio at expiry is
$$\max(S_T-K,0) + K = \max(S_T,K)$$
and therefore the portfolio's payoff is at least as much as the stock, and thus it should always be worth at
least as much at all times. Hence
$$c + K B_0 \geq S_0$$
which is equivalent to
$$c \geq S_0 - K B_0$$
If we assume positive interest rates, then $B_0 < 1$ and hence we have
$$S_0-K < S_0 - K B_0 \leq c \leq C \leq S_0$$
:eqlabel:`callbounds`

## It is never optimal to exercise an American call on a non-dividend paying stock

We can now deduce that for a non-dividend paying stock, an American call should never be exercised before expiry,
because it is always worth more than the intrinsic value. Also, we can deduce that the American call has the same
value as the European call for a non-dividend paying stock since it is never optimal to exercise it before expiry.

We have shown earlier that the American call price is an increasing function of the maturity, but that it was not 
necessarily the case of European calls. Now, we can say that for European calls on non-dividend paying stocks the relation holds given that European calls on non-dividend paying stocks have the same value as the corresponding American calls.


## The call price is continuous and mildly decreasing as a function of the strike

We now show that the difference in price of two European call options at different strikes but the same expiry is
less than the strike difference. Let $c_1$ and $c_2$ be two calls with expiry $T$ and strikes $K_1 < K_2$ respectively. We know that $c_1$ is always worth more than $c_2$. On the other hand a portfolio of $c_2$ and $K_2 - K_1$ zero-coupon bonds will pay at maturity $T$
$$(K_2-K_1)+\max(S-K_2,0)$$
That portfolio's payoff is larger than the payoff of $c_1$
$$(K_2-K_1)+\max(S-K_2,0) \geq \max(S-K_1,0)$$
and at all times the value of the portfolio is
$$c_2(t)+(K_2-K_1)B_t$$
In summary we have
$$c_2(t) \leq c_1(t) \leq c_2(t) + (K_2-K_1)B_t$$
and thus
$$0 \leq c_1(t)-c_2(t) \leq (K_2-K_1)B_t$$
If we let $K_2$ tend to $K_1$, then the price of $c_2$ must converge to that of $c_1$, and hence the price of an 
option is a continuous function of the strike. 

The inequalities are also equivalent to
$$0 \leq \frac{c_1(t)-c_2(t)}{K_2-K_1} \leq B_t$$
and
$$-B_t \leq \frac{c_2(t)-c_1(t)}{K_2-K_1} \leq 0$$
and taking the limit when $K_2 \rightarrow K_1$ we have that
$$-B_t \leq \lim_{K_2 \rightarrow K_1}\frac{c_2(t)-c_1(t)}{K_2-K_1} = \frac{\partial c}{\partial K} \leq 0$$
and hence the derivative of the call price with respect to the strike is negative as expected given we have already 
shown that the call option is a decreasing function of the strike. The new finding is that the derivative with respect
to the strike is bounded below by $-B_t$ and given $-1 \leq -B_t \leq0$, we can conclude that not only the call price
is a continuous decreasing function of the strike, but that the prices cannot decrease too rapidly.

For American calls on dividend paying stocks, we only have the weaker inequalities
$$0 \leq C_1(t)-C_2(t) \leq (K_2-K_1)$$
and the derivative inequalities
$$-1 \leq  \frac{\partial C}{\partial K} \leq 0$$

## Option prices are convex functions of strike

Given a parameter $0\leq \theta \leq 1$, consider the portfolio of 

- $\theta$ call options with strike $K_1$, 
- $1-\theta$ call options with strike $K_2$, and
- $-1$ call option with strike $K_3=\theta K_1 +(1-\theta)K_2$
  
Given the call option payoff function is convex, at expiry we have
$$\theta \max\left[S-K_1,0\right] + (1-\theta) \max\left[S-K_2,0\right] \geq \max\left[S-\theta K_1-(1-\theta)K_2,0\right]$$
By arbitrage, the value of the portfolio at any time satisfies
$$\theta c(K_1) + (1-\theta) c(K_2) \geq c(\theta K_1+(1-\theta)K_2)$$
and hence the value of the portfolio must always be non-negative. Thus, we have shown
that the option price is a __convex__ function of the its exercise price
$$\theta c(K_1) + (1-\theta) c(K_2) \geq c(\theta K_1+(1-\theta)K_2) \forall 0 \leq \theta \leq 1$$

## Option prices are homogenous functions of degree one

The call option price function is homogeneous of degree one in $S$ and $K$ with respect to scale:
$$C(\alpha S, \alpha K) = \alpha C(S,K)$$
and
$$c(\alpha S, \alpha K) = \alpha c(S,K)$$
This stems from the payoff function
$$\max\left[\alpha S- \alpha K,0\right]=\max\left[\alpha (S-K),0\right]=\alpha \max\left[ S-K,0\right]$$
So the option to buy $\alpha$ stocks at a strike price of $\alpha K$ must be worth the same as $\alpha$ calls to
buy one stock at a strike price of $K$.

If we take $\alpha= \frac{1}{K}$, then we have
$$\max\left[S - K,0\right]= K\max\left[\frac{S}{K}-1,0\right]$$
and we have 
$$c(S,K)=K c(\frac{S}{K}, 1)$$
These relationships hold for American and European calls and puts.
