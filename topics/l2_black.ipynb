## First look at the Profit and Loss of a delta-hedged option

In [Black and Scholes (1973)](../readings/r5_blackscholes.html), we saw that the change 
in value of a portfolio with a delta-hedged option $\pi = \Delta S + C$ is given by
$$d\pi = \Theta dt + \frac{1}{2} \Gamma \sigma^2 S^2 dt$$
where
$\Theta = \frac{\partial C}{\partial t}$ and $\Gamma = \frac{\partial^2 C}{\partial S^2}$.

The second term is quadratic in $dS$ and much smaller than the linear change in value of $\pi$ 
proportional to $dS$ which has been canceled out by the delta hedging. We have also seen in 
previous lectures that options have positive convexity, i.e. the $\Gamma$ is positive. Hence,
a delta-hedged option benefits from the convexity or curvature of the option.

Given the risk has been removed by the delta-hedging, we also saw that means the portfolio
should earn the risk-free rate and thus:
$$\Theta + \frac{1}{2} \Gamma \sigma^2 S^2 = r \pi$$

Let's ignore the interest rates for now, i.e. assume $r=0$ for the rest of this section.
Then
$$\Theta = -\frac{1}{2} \Gamma \sigma^2 S^2$$
and so $\Theta dt$ is the time decay that offsets the gain from convexity as the option gets closer to expiry.

In the Black-Scholes formula, all the inputs are known except for the volatility $\sigma$. 
The value of $\sigma$ that matches the model output to market prices is called the implied volatility
which we denote $\Sigma$.
It is the value that the __unknown future__ volatility has to be in order that the model will have
valued the option correctly ex-ante. We could say that the implied volatility is the market's expected
value of future volatility.

If the future realized $\sigma$ ends up greater that the implied value $\Sigma$, the delta-hedged option
will increase in value more than expected. Similarly, if the $\sigma$ ends up lower that $\Sigma$,
then the delta-hedged option will appreciate less than anticipated.

The amount we expect to lose due to time decay during $dt$ is 
$$\frac{1}{2} \Gamma \Sigma^2 S^2 dt$$
while the gain from convexity if the stock is actually moving with a volatity of $\sigma$ is
$$\frac{1}{2} \Gamma \sigma^2 S^2 dt$$
and thus the infinitesimal profit or loss after time $dt$ is given by
$$\frac{1}{2} \Gamma S^2 (\sigma^2-\Sigma^2) dt$$
This shows that when we delta-hedge an option, we are in effect take a position on volatility:

- a long position in the delta-hedged option profits from higher realized volatility
- a short position in the delta-hedged option profits from lower realized volatility
