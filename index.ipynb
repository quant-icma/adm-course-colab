## Overview

This course provides a solid introduction to derivatives modelling, blending theoretical and mathematical derivations with programming implementation. It is a relatively advanced introduction. It will not make you an expert on the subject but it does go beyond the Black-Scholes-Merton model, exploring implications of loosening the various assumptions that underlie the Black-Scholes-Merton model and theory. The course assumes you have taken at least a one-semester course on stochastic calculus. It also assumes that you have studied (or are currently studying) an introductory course on Black-Scholes-Merton - the Derivatives Modelling course at the ICMA centre - pricing of vanilla European options.

## Learning Objectives

By taking this course, you expect to:

  - acquire a good understanding of arbitrage pricing and replication theory and practice,
  - get some exposure to derivatives modelling beyond Black-Scholes-Merton model,
  - learn about modelling derivatives beyond the Vanilla European Calls and Puts
  - practice implementing derivatives models, i.e. get your 'hands dirty' at coding model,
  - gain insights into assumptions and limitations of models and their scope of usefulness.

## Prerequisites

  - Stochastic Calculus
  - Derivatives Modelling
  - Highly recommended to also take ICM-177 Programming for Financial Engineering

## Course Format

The course consists of:

  - 1 live introduction session of 1 hour with the lecturer
  - 10 live interactive lectures of 1 hour with the lecturer
  - 10 pre-recorded lectures of 1 hour and 30 minutes taught by the lecturer
  - A number of 'seminal' papers readings
  - 4 live interactive seminars of 1 hour with the TA
  - 4 pre-recorded seminars of 1 hour taught by the TA  
  - 4 graded group assignments
  - 1 live exam review session of 1 hour with the lecturer
  - 1 live recap and conclusion session of 1 hour with the lecturer
